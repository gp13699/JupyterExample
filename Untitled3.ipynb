```python
# Quick utility to embed the videos below
from IPython.display import YouTubeVideo
def embed_video(index, playlist='PLYCpMb24GpOC704uO9svUrihl-HY1tTJJ'):
    return YouTubeVideo('', index=index - 1, list=playlist, width=600, height=350)
```

```python
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL,'fremont.csv')
```

```python
import pandas as pd
data = pd.read_csv('fremont.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
embed_video(2)
```

```python
plt.style.use('seaborn')
data.columns = ('W', 'E')
data.resample('W').sum().plot();
```

```python
data['Total'] = data['W'] + data['E']
ax=data.resample('D').rolling(356).sum().plot();
ax.set_ylim(0,None);
```

```python
ax1=data.groupby(data.index.time).mean().plot();
```

```python
pivoted = data.pivot_table('Total' ,index=data.index.time ,columns=data.index.date)
pivoted.iloc[:5, :5]
```

```python
pivoted.plot(legend=False, alpha=0.02);
```
